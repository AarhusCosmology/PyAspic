```python
import numpy as np
import matplotlib.pyplot as plt
import pyaspic
from Slowroll import Slowroll

lnRreh = np.linspace(-45,10,10)
alpha = np.linspace(np.sqrt(2)*10**(-5/2),0.9,100)
f = np.linspace(3,30,100)
```

```python
pli_ns = []
pli_r = []
pli_lnRreh = []

ni_ns = []
ni_r = []
ni_lnRreh = []

for i in lnRreh:
    for j in alpha:
        pli = Slowroll('pli',i,j)
        pli_ns.append(pli.ns)
        pli_r.append(pli.r)
        pli_lnRreh.append(i)
    for j in f:
        ni = Slowroll('ni',i,j)
        ni_ns.append(ni.ns)
        ni_r.append(ni.r)
        ni_lnRreh.append(i)
        
```

```python
pli = plt.scatter(pli_ns,pli_r,c=pli_lnRreh)
ni = plt.scatter(ni_ns,ni_r,c=ni_lnRreh)
plt.xlim([0.83,1.03])
plt.ylim([0.00,0.45])
plt.xlabel('n_s')
plt.ylabel('r')
cbar = plt.colorbar()
cbar.set_label('ln(R_reh)', rotation=270)
legend1 = plt.legend(['Power law inflation'], loc='upper right')
legend2 = plt.legend(['Natural inflation'], loc='lower left')
plt.gca().add_artist(legend1)
```
